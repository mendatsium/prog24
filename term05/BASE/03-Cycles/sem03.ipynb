#### Задача 1.

Пользователь вводит какие-то строки, которые наша программа должна приводить в "заголовочный" вид: все слова в строке с заглавной буквы. Когда пользователь вводит "стоп", программа должна завершить работу, причем ей неважно, в каком регистре введен "стоп", хоть заборчиком.

```python
s = input().title()
while s != 'Стоп':
    print(s)
    s = input().title()
```

#### Задача 2.

Соберем, кстати, заборчик: пользователь вводит какое-то слово маленькими буквами, а мы должны его собрать таким образом, чтобы каждая вторая буква была заглавной.

```python
s = input()
result = ''
for i in range(len(s)):
    if i % 2:
        result += s[i].upper()
    else:
        result += s[i]
print(result)
```

#### Задача 3.

#### Задача 4.

Пользователь вводит целое число, в котором заранее мы не знаем, сколько цифр. Программа должна вернуть сумму цифр этого числа. 

```python
num = input()
res = 0

for i in num:
    if i.isdigit():
        res += int(i)
print(res)
```

#### Задача 6.

Напишем программу для проверки подобранных рифм: пользователь в цикле вводит два слова, программа должна проверить, что у них совпадает не меньше четырех последних букв (будем наивно считать, что все читается, как пишется). Цикл останавливается, если первое из введенных слов не содержит букв. 

```python
word1, word2 = input().lower(), input().lower()

while word1.isalpha():
    if len(word1) < 4 or len(word2) < 4:
        print('Слишком коротко!')
    elif word1[-1:-5:-1] != word2[-1:-5:-1]:
        print('Слова не рифмуются')
    else:
        print('Слова рифмуются')
    word1, word2 = input().lower(), input().lower()
```

#### Задача 7. 

Напишем программу-игру: наша программа будет загадывать случайное число в интервале от 1 до 1024, а потом в цикле предлагать игроку его угадать. Если игрок угадал, то игра завершается победой, если назвал неправильное число, программа должна сообщать, больше это число или меньше, чем загаданное. Всего дается 10 попыток отгадать, если они все закончились, игра проиграна!

```python
from random import randint 

secret_number = randint(1, 1024) # число мы загадали, осталось написать остальное

for i in range(1, 11):
    num = int(input())
    if num == secret_number:
        print('Угадано')
        break
    elif num > secret_number:
        print('Число больше загаданного')
    else:
        print('Число меньше загаданного')
else:
    print('Вы проиграли :(')
```

#### Задача 9. 

Маша с математического факультета МГУ сильно заскучала, идя по улице. Она решила придумать себе такое развлечение:

Если она видит машину, она проводит арифметические операции с ее номером. Так, если ее номер делится на 3, она возводит результат во 2-ю степень. Если ее номер не делится на 3, она целочисленно делит его на 5.

Если она видит магазин, она читает его название наоборот.

Если Маша видит группу человек, она умножает количество людей на 8 и возвращает остаток от деления на 6.

С помощью input() пользователь вводит либо целое трехзначное число (номер машины), либо строку (название магазина, не короче четырех букв), либо однозначное или двузначное целое число (количество людей).

```python
s = input()

if s.isdigit() and len(s) == 3:
    s = int(s)
    if int(s) % 3 == 0:
        print(s ** 2)
    else:
        print(s // 5)

elif len(s) >= 4:
    print(s[::-1])
    
else: 
    s = int(s)
    print((s * 8) % 6)
    
```
