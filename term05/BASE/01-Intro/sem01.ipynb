#### Задача 3. 

Пользователь вводит число (int). Если число четное, разделите его целочисленным делением на 4. Если число нечетное, верните остаток от деления. Результат выведите с помощью print().

```python
number = int(input())
if number % 2:  #если процент от деления True, т.е. 1 => нечетное
    print(number % 4)
else:   #четное
    print(number // 4)
```

#### Задача 6. 

Как известно, у квадратного уравнения ($ax^2 + bx + c$) либо нет корней вообще, либо есть один корень, либо два. Пользователь вводит коэффициенты a, b, c: ваша программа должна понять, сколько у этого уравнения корней, и вычислить их, если есть хотя бы один. (Погуглите про дискриминант!)

```python
a = int(input())
b = int(input())
c = int(input())

D = b ** 2 - 4 * a * c 

if D < 0:
    print('Нет корней.')
elif D == 0:
    print(-b / 2 * a)
else:
    print((-b + D ** 0.5) / 2 * a, (-b - D ** 0.5) / 2 * a)
```

#### Задача 7. 

Пользователь вводит количество минут. Это число - время до выхода Миши на концерт (на сборы и поездку). Задача - напечатать действия (по одной строке в результате проверки условий), которые Миша точно успеет сделать (с учетом времени, которое требуется на дорогу). Действия: Принять ванну (2 часа 10 минут), порепетировать (1 час 20 минут), поменять струны (15 минут), побриться (7 минут). Поездка до клуба займет 1 час 5 минут. Если Мише не хватает времени ни на одно дело, выведите "Миша поедет неподготовленным". Если Мише не хватает времени даже на дорогу, выведите "Миша опоздал на концерт".

```python
t = int(input())

if t < 65:
    print('Миша опоздал на концерт')
elif t < (65 + 7):
    print('Миша поедет неподготовленным')
if t >= (65 + 7):
    print('Миша успеет побриться')
if t >= (65 + 7 + 15):
    print('Миша успеет поменять струны')
if t >= (65 + 7 + 15 + 80):
    print('Миша успеет порепетировать')
if t >= (65 + 7 + 15 + 80 + 130):
    print('Миша успеет принять ванну')
```

#### Задача 8.

Пользователь вводит, сколько часов по времени в Москве (счет времени 24). Ивана срочно вызвали в командировку, и Маша не успела даже спросить, куда он едет. Спустя некоторое время от него приходит сообщение: "У нас час ночи, я пошел спать". Маша смотрит на часы и видит введенное пользователем количество часов. Маша догадывается, что Ивана могли отправить в столицу одной из этих стран: Америка, Япония, Китай, Бразилия или Испания. Помогите Маше узнать, в какой стране сейчас Иван. Если ее догадки насчет возможных стран ложны, выведите "Маша не знает, где Иван".

```python
t = int(input('Введите время на часах у Маши (0-24)'))

if t > 24 or t < 0:
    print('Лох, не бывает столько часов')
elif t == 8:
    print('Иван в Вашингтоне (столица Америки)')
elif t == 19:
    print('Иван в Токио (столица Японии)')
elif t == 20:
    print('Иван в Пекине (столица Китая)')
elif t == 7:
    print('Иван в Бразилиа (столица Бразилии)')
elif t == 3:
    print('Иван в Мадриде (столица Испании)')
else:
    print('Маша не знает, где Иван')
```
