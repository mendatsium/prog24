#### Задача 1. 

Напишите функцию, которая будет принимать числа от 1 до 10 и заменять их на слова: 1 = один, 2 = два и т.п. 

```python
def nums_cyrillic(user_num):
    nums_cyr = ['один', 'два', 'три', 'четыре', 'пять', 'шесть', 'семь', 'восемь', 'девять', 'десять']
    dict_nums = {x + 1: y for x, y in enumerate(nums_cyr)}
    for k, v in dict_nums.items():
        if user_num == k:
            print(f'{k} = {v}')

user_num = int(input())
nums_cyrillic(user_num)
```

#### Задача 2. 

Пользователь вводит две строки (в каждой строке какое-то слово). Напишите функцию, которая будет проверять, что из букв строки 1 можно составить слово в строке 2, причем количество букв тоже нужно учитывать: нельзя составить слово "мама" из слова "марокко". Функция должна возвращать True или False. Мы делали такую задачу с помощью методов строк, а теперь давайте попробуем с помощью словарей.

```python
def words_eq(s1, s2):
    dict1 = {}
    for symb in s1:
        dict1[symb] = s1.count(symb)
    good = True
    for symb in s2:
        if s2.count(symb) > dict1[symb]:
            good = False
    return good

s1 = input()
s2 = input()
words_eq(s1, s2)
```

#### Задача 3.

В одном необычном ресторане все блюда добавляются в меню только после того, как их закажут клиенты. Но ресторан не успевает записывать все заказы и просит вас помочь.

1. Напишите программу, которая будет:
   - Собирать заказы клиентов в виде словаря, где ключ — это название блюда, а значение — количество заказов этого блюда.
   - Позволять клиентам добавлять новые заказы.
   - Удалять блюдо, если его заказали по ошибке.
   - Показывать текущее меню и количество заказов каждого блюда.

2. Реализуйте следующую функциональность:
   - Если клиент заказывает уже существующее блюдо, программа должна увеличивать количество заказов.
   - Если клиент хочет отменить заказ блюда, программа должна уменьшать количество заказов. Если количество заказов становится равным нулю, блюдо должно удаляться из меню.
   - Программа завершает работу, если клиент вводит команду выход.
   - Добавьте проверку на ввод: нельзя вводить отрицательные количества.
   - Добавьте возможность изменить количество заказов, указав новое число, вместо добавления или удаления.

3. Пример работы:
  
       Добро пожаловать в "Ресторан приключений"!
       Доступные команды: добавить, удалить, меню, выход

       Введите команду: добавить
       Введите название блюда: Пицца
       Введите количество: 2
       Блюдо "Пицца" добавлено в меню с количеством 2.

       Введите команду: добавить
       Введите название блюда: Пицца
       Введите количество: 1
       Блюдо "Пицца" обновлено: теперь количество 3.

       Введите команду: меню
       Текущее меню:
       Пицца: 3

       Введите команду: удалить
       Введите название блюда: Пицца
       Введите количество: 3
       Блюдо "Пицца" удалено из меню.

       Введите команду: выход
       До свидания!
   

```python
print('Добро пожаловать в "Ресторан приключений"!', '\n', 
    'Доступные команды: добавить, удалить, меню, выход')

menu = {}
command = input('введите команду:').lower()

while command != 'выход':
    if command == 'добавить':
        meal = input('Введите название блюда:').lower()
        num = int(input('Введите количество:'))
        if meal not in menu:
            menu[meal] = num
            print(f'Блюдо {meal} добавлено в ваше меню в количестве {menu[meal]}.')
        else:
            menu[meal] += num
            print(f'Блюдо {meal} обновлено: теперь количество {menu[meal]}.')
        # print(menu)
        command = input('введите команду:').lower()
    elif command == 'удалить':
        meal = input('Введите название блюда:').lower()
        num = int(input('Введите количество:'))
        if meal not in menu:
            print('Такого блюда нет в вашем меню.')
        elif menu[meal] < num:
            print(f'Невозможно удалить блюдо {meal} в таком количестве: этого блюда в вашем меню всего {menu[meal]}')
        elif menu[meal] == num:
            del menu[meal]
            print(f'Блюдо {meal} удалено из вашего меню')
        else:
            menu[meal] -= num
            print(f'Блюдо {meal} обновлено: теперь количество {menu[meal]}.')
        # print(menu)
        command = input('введите команду:').lower()
    elif command == 'меню':
        print('Текущее меню:')
        for k, v in menu.items():
            print(f'{k}: {v}')
        command = input('введите команду:').lower()
else:
    print('До свидания!')
```
