#### Задача 1. 

Выберите любую картинку и с помощью библиотеки Pillow

- примените к ней любой фильтр
- отразите ее горизонтально (погуглите, как)
- уменьшите ее до размера 32х32 пикселя
- сохраните ее в формате ico

```python
from PIL import Image, ImageFilter

try:  
    original = Image.open('dagestan.jpg')  
except FileNotFoundError:  
    print("Файл не найден")

smooth = original.filter(ImageFilter.SMOOTH)
converted_img = smooth.transpose(Image.FLIP_LEFT_RIGHT)
converted_img.thumbnail((32,32))
converted_img.save('dagestan.ico')
```

#### Задача 3. 

Напишите скрипт, который будет конвертировать любые картинки формата .jpg (расширение может быть написано несколькими разными способами) в .png. 

```python
from os import path

route = 'dagestan.jpg'

if not route.lower().endswith('.jpg') or route.lower().endswith('.jpeg'):
    print("Это не те дроиды, которых вы ищите")
else:
    name, _ = path.splitext(route)
    try:
        Image.open(route).save(name + '.png')  
    except FileNotFoundError:  
        print("Файл не найден")
```

#### Задача 4.

Сделайте интерактивную карту Москвы и разместите на ней маркер для РГГУ.

```python
import folium

moscow_map = folium.Map(location=[55.7522, 37.6156], zoom_start=11)
folium.Marker([55.7777, 37.5964],
              popup='Российский Государственный Гуманитарный Университет',
              tooltip='РГГУ'
             ).add_to(moscow_map)
moscow_map
```

#### Задача 5. 

Выберите любой признак на WALS и перенесите карту для него оттуда в python с помощью folium.

```python
# Feature 67A: The Future Tense

import folium
import json 

small = json.load(open('fut_ex.geojson'))
averange = json.load(open('no_fut.geojson'))

map = folium.Map(location=[55, 100], zoom_start=3)

folium.GeoJson(small, zoom_on_click=True, marker=folium.Marker(icon=folium.Icon(color='blue', icon='check', prefix='fa'))).add_to(map)
folium.GeoJson(averange, zoom_on_click=True, marker=folium.Marker(icon=folium.Icon(color='red', icon='ban', prefix='fa'))).add_to(map)

map
```
